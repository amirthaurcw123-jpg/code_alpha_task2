```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/Unemployment in India.csv")
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.columns = df.columns.str.strip()
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.head()
```

```python
df['Estimated Unemployment Rate (%)'].mean()
```

```python
state=df.groupby('Region')['Estimated Unemployment Rate (%)'].mean()
plt.figure(figsize=(12,6))
state.plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Average Unemployment Rate by Region')

```

```python
monthly=df.groupby(df['Date'].dt.month)['Estimated Unemployment Rate (%)'].mean()
plt.figure(figsize=(8,8))
monthly.plot(kind='bar')
plt.xlabel('Month')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Average Unemployment Rate by Month')
```

```python
before=df[df['Date']<'2020-03-01']
print(before['Estimated Unemployment Rate (%)'].mean())
```

```python
after=df[df['Date']>='2020-03-01']
print(after['Estimated Unemployment Rate (%)'].mean())
```

```python
covid=df.groupby('Date')['Estimated Unemployment Rate (%)'].mean()
plt.figure(figsize=(12,6))
covid.plot(kind='bar')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate Over Time')
```

```python
df['Month']=df['Date'].dt.month
season=df.groupby('Month')['Estimated Unemployment Rate (%)'].mean()
plt.figure(figsize=(8,8))
season.plot(kind='bar')
plt.xlabel('Month')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Average Unemployment Rate by Season')
```

**Insights and Policy Recommendations**
*   The analysis shows that unemployment increased significantly during the Covid-19 pandemic.
*   Some regions consistently recorded higher unemployment rates than others.
*   Monthly trends indicate that unemployment rates varied over time due to economic conditions.
*  The government can reduce unemployment by creating more job opportunities, promoting skill development programs, and supporting small and medium-sized businesses.
*  During economic crises, employment guarantee schemes and financial assistance can help reduce the impact of unemployment.
*  Continuous monitoring of unemployment trends can help policymakers make informed economic and social decisions.
